## Daily Coding Problem: Problem #98 [Easy]

This problem was asked by Coursera.

Given a 2D board of characters and a word, find if the word exists in the grid.

The word can be constructed from letters of sequentially adjacent cell, where "adjacent" cells are those horizontally or vertically neighboring. The same letter cell may not be used more than once.

For example, given the following board:

[
  ['A','B','C','E'], <br>
  ['S','F','C','S'], <br>
  ['A','D','E','E']] <br> <br>
exists(board, "ABCCED") returns true, <br>exists(board, "SEE") returns true, <br>exists(board, "ABDE") returns false.

```python
import pandas as pd

board1 = [['A', 'B', 'C', 'E'],
          ['S', 'F', 'C', 'S'],
          ['A', 'D', 'E', 'E']]
board1 = pd.DataFrame(board1)


def exists(board, string):
    
    # instantiating object with starting points equal to first letter of string
    global starting_points
    starting_points = []
    for i in range(len(board.index)):
        for j in range(len(board.columns)):
            if board.iloc[i, j] == string[0]:
                starting_points.append([i, j])

    # helper function to set floor ar zero (needed when navigating through board to avoid jumping from one end to another (subsetting with negative indexing)
    def zero_floor(value):
        if value >= 0:
            return value
        elif value < 0:
            return 100 # can cause trouble when dataframe has more than 100 rows or columns
        else:
            print("Error on zero_floor")

    # helper function checking if right neighbor is equal to next string element. If so, return coordinates of next position
    def check_right_neighbor(coord, string):
        try:
            new_row = coord[0]
            new_col = coord[1] + 1
            if board.iloc[new_row, new_col] == string:
                next_point = [new_row, new_col]
                return next_point
        except:
            pass

    # helper function checking if left neighbor is equal to next string element. If so, return coordinates of next position
    def check_left_neighbor(coord, string):
        try:
            new_row = coord[0]
            new_col = zero_floor(coord[1] - 1)  # zero floor
            if board.iloc[new_row, new_col] == string:
                next_point = [new_row, new_col]
                return next_point
        except:
            pass

    # helper function checking if below neighbor is equal to next string element. If so, return coordinates of next position
    def check_below_neighbor(coord, string):
        try:
            new_row = coord[0] + 1
            new_col = coord[1]
            if board.iloc[new_row, new_col] == string:
                next_point = [new_row, new_col]
                return next_point
        except:
            pass

    # helper function checking if above neighbor is equal to next string element. If so, return coordinates of next position
    def check_above_neighbor(coord, string):
        try:
            new_row = zero_floor(coord[0] - 1)  # zero floor
            new_col = coord[1]
            if board.iloc[new_row, new_col] == string:
                next_point = [new_row, new_col]
                return next_point
        except:
            pass

    # helper function defining next coordinates of possible path (see known bugs)
    def find_next_point(coord, string):
        next_point_list = []
        next_point_list.append(check_right_neighbor(coord=coord, string=string))
        next_point_list.append(check_left_neighbor(coord=coord, string=string))
        next_point_list.append(check_below_neighbor(coord=coord, string=string))
        next_point_list.append(check_above_neighbor(coord=coord, string=string))
        next_point_list = [x for x in next_point_list if x is not None]
        return next_point_list

    try:
        result_list1 = []
        
        # iterating through all possible starting points
        for iter0 in range(len(starting_points)):
            result_list2 = []
            global point
            point = []
            intermediate = starting_points[iter0]
            point.append(intermediate) # converting coordinates to list of lists
            
            # iterating over each element in string
            for iter1 in range(1, len(string)):
                
                # checking if there is a next point on path. If so, append 'true' to result_list2 and continue. If not, append 'false'to result_list2 and break
                point = find_next_point(coord=point[0], string=string[iter1])
                if point != []:
                    result_list2.append("true")
                elif point == []:
                    result_list2.append("false")
                    break
            
            # if one false connection is in result_list2, the entire path yields 'false' to be appended to result_list1
            if "false" in result_list2:  
                result_list1.append("false")
            else:
                result_list1.append("true")
        
        # assessing result_list1. If one path (checked with result_list1) was 'true' there is a path with current string and output of overall function is 'true'
        if "true" in result_list1:
            print("true")
        else:
            print("false")
    except:
        print("false")


exists(board = board1, string = "ABCCED")
exists(board = board1, string = "SEE")
exists(board = board1, string = "ABDE")

# Known bugs:
## In cases such as board1[1,3] checking for 'E' where two paths are possible to move forward, the order of components (right, left, below, above) in the find_next_point() function matters. Only the first path (right, left, below, above) after the split will be checked
## Positions/coordinates can be re-used. If question is extended to NOT re-using elements, code needs to be extended to temporarily delete used positions
```
